```python
# environment
import imp
import os, sys

# standard libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# k-Means clustering algorithm libraries
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale

```

```python
%matplotlib inline
sns.set_theme()
```

```python
fp = 'cluster_raw.dat'
```

```python
# after reviewing data, it has headers, tab delimiter, index is first column
df = pd.read_csv(fp, sep='\t', index_col=0)
```

```python
df.head()
```

```python
# examine
df.info()
```

```python
# examine relationships
sns.pairplot(df);
```

```python
# prepare x
Xm = df[ ['x0','x1','x2','x3','x4'] ]
```

```python
# scale all since they are not of the same scale
data = scale(Xm)
```

```python
# apply kMeans clustering
KMeans(init='k-means++', n_clusters=2, n_init=10)
```

```python
kmeans = KMeans(n_clusters=4, random_state=0).fit(data)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
data
```

```python
data.shape
```

```python
y_kmeans = kmeans.predict(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
